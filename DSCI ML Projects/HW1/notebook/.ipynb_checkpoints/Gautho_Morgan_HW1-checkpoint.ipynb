```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn import neighbors
import warnings
warnings.filterwarnings('ignore')
```

#### (a) Download the Vertebral Column Data Set from: https://archive.ics.uci.edu/ml/datasets/Vertebral+Column

```python
data = pd.read_csv("../data/column_2C.dat",sep=" ", header=None, names=["pelvic_incidence","pelvic_tilt","lumbar_lordosis_angle","sacral_slope","pelvic_radius","grade_of_spondyolisthesis","class"])
classVal = np.zeros((np.shape(data)[0],1))
for index, row in data.iterrows():
    classVal[index,0] = (row["class"]=="AB")
data = data.assign(classVal=classVal)
display(data)
```

ii. Make boxplots for each of the independent variables. Use color to show

iii. Select the first 70 rows of Class 0 and the first 140 rows of Class 1 as the
training set and the rest of the data as the test set.

```python
c0 = data[data["class"]=="NO"]
c1 = data[data["class"]=="AB"]
training_set = pd.concat([c0.head(70), c1.head(140)])
test_set = pd.concat([c0[70:], c1[140:]])
```

#### (c) Classification using KNN on Vertebral Column Data Set

i. Write code for k-nearest neighbors with Euclidean metric (or use a software
package).


```python
def knnClassifier(k):
    return neighbors.KNeighborsClassifier(k, metric = "euclidean")
```

ii. Test all the data in the test database with k nearest neighbors. Take decisions by majority polling. Plot train and test errors in terms of k for
k âˆˆ {208, 205, . . . , 7, 4, 1, } (in reverse order). You are welcome to use smaller
increments of k.

```python
k_vals = np.flip(np.arange(1,209,3))
train_error = []
test_error = []
for k in k_vals:
    trainModel = knnClassifier(k)
    trainModel = trainModel.fit(training_set.iloc[:,0:6], training_set.iloc[:,7])
    trainPreds = trainModel.predict(training_set.iloc[:,0:6])
    testPreds = trainModel.predict(test_set.iloc[:,0:6])
    train_error.append(1 - metrics.classification_report(trainPreds, training_set.iloc[:,7], output_dict = True).get('accuracy'))
    test_error.append(1 - metrics.classification_report(testPreds, test_set.iloc[:,7], output_dict = True).get('accuracy'))


    

# Plot
error = pd.DataFrame({"k":k_vals, "train_error":train_error,"test_error":test_error})
plt.plot(error["k"], error["train_error"], label = "Training Error")
plt.plot(error["k"], error["test_error"], label = "Test Error")
plt.gca().invert_xaxis()
plt.xlabel("k value")
plt.ylabel("Error Rate")
plt.title("Test vs. Training Error Rate")
plt.legend()
plt.show()


# Needed for part f
train_dict = []
train_k_star = error[error["train_error"] == error["train_error"].min()]
train_dict.append({"metric":"euclidian","k": train_k_star.iloc[0]["k"], "train_error":train_k_star.iloc[0]["train_error"]})
```

ii. (cont) Which kâˆ— is the most suitable k among those values?

```python
k_star = error[error["test_error"] == error["test_error"].min()]
print("K*, or the most suitable value of k, is %d with a test error rate of %.2f" % (k_star.iloc[0]["k"],k_star.iloc[0]["test_error"] ), "%.")
```

ii. (cont) Calculate the confusion matrix, true positive rate, true negative rate, precision,
and F1-score when k = k*.

```python
# retrain with best k

trainModel = knnClassifier(int(k_star.iloc[0]["k"]))
trainModel = trainModel.fit(training_set.iloc[:,0:6], training_set.iloc[:,7])
trainPreds = trainModel.predict(training_set.iloc[:,0:6])
testPreds = trainModel.predict(test_set.iloc[:,0:6])

conf_mat = metrics.confusion_matrix(test_set.iloc[:,7],testPreds)
tn, fp, fn, tp = conf_mat.ravel()
sns.heatmap(conf_mat,square= True, annot=True, cbar= False)
plt.xlabel("Predicted Value")
plt.ylabel("True Value")
plt.title("Confusion Matrix for k=k*")
plt.show()

tpr = tp / (tp + fn)
tnr = tn / (tn + fp)
prec = tp / (tp + fp)
recall = tp / (tp + fn)
f1score = 2 * (prec * recall) / (prec + recall)
print("The true positive rate is %f.\nThe true negative rate is %f.\nThe precision is %f.\nThe F1-score is %f." % (tpr, tnr, prec, f1score))

```

iii. Since the computation time depends on the size of the training set, one may only use a subset of the training set. Plot the best test error rate, which is obtained by some value of k, against the size of training set, when the
size of training set is N âˆˆ {10, 20, 30, . . . , 210}. Note: for each N, select
your training set by choosing the first [N/3] rows of Class 0 and the first
N âˆ’ [N/3] rows of Class 1 in the training set you created in 1(b)iii. Also, for
each N, select the optimal k from a set starting from k = 1, increasing by 5.
For example, if N = 200, the optimal k is selected from {1, 6, 11, . . . , 196}.
This plot is called a Learning Curve.

```python
n_vals = np.arange(10,211,10)

k_opt = []
best_error = []

for n in n_vals:
    train_set_n = pd.concat([c0.head(round(n/3)), c1.head(n - round((n/3)))])
    k_vals_n = np.arange(1,n,5)
    min_error_n = 100
    min_error_k = 0
    for k in k_vals_n:
        trainModel = knnClassifier(k)
        trainModel = trainModel.fit(train_set_n.iloc[:,0:6], train_set_n.iloc[:,7])
        trainPreds = trainModel.predict(train_set_n.iloc[:,0:6])
        testPreds = trainModel.predict(test_set.iloc[:,0:6])
        k_error = 1 - metrics.classification_report(testPreds, test_set.iloc[:,7], output_dict = True).get('accuracy')
        if(k == 1):
            min_error_n = k_error
            min_error_k = k
        elif(min_error_n > k_error):
            min_error_n = k_error
            min_error_k = k
    best_error.append(min_error_n)
    k_opt.append(min_error_k)
    
# Plot
plt.plot(n_vals, best_error, label = "Best Test Error")
plt.xlabel("Number of training values (N)")
plt.ylabel("Best Test Error Rate")
plt.title("Best Test Error vs. Number of Training Examples")
plt.legend()
plt.show() 
```

#### (d) Replace the Euclidean metric with the following metrics and test them. Summarize the test errors (i.e., when k = kâˆ—) in a table. Use all of your training data and select the best k when {1, 6, 11, . . . , 196}.


```python
metric_dict=[]
training_x = training_set.iloc[:,0:6]
training_y = training_set.iloc[:,7]
test_x = test_set.iloc[:,0:6]
test_y = test_set.iloc[:,7]
```

i. Minkowski Distance:

A. which becomes Manhattan Distance with p = 1.

```python
k_vals = np.arange(1,197,5)
manhattan_train_error=[]
manhattan_test_error = []
for k in k_vals:
    trainModel = neighbors.KNeighborsClassifier(k, metric = "manhattan")
    trainModel = trainModel.fit(training_x, training_y)
    trainPreds = trainModel.predict(training_x)
    testPreds = trainModel.predict(test_x)
    manhattan_train_error.append(1 - metrics.classification_report(trainPreds, training_y, output_dict = True).get('accuracy'))
    manhattan_test_error.append(1 - metrics.classification_report(testPreds, test_y, output_dict = True).get('accuracy'))


    
manhattan_error = pd.DataFrame({"k":k_vals, "train_error":manhattan_train_error,"test_error":manhattan_test_error})

k_star = manhattan_error[manhattan_error["test_error"] == manhattan_error["test_error"].min()]
metric_dict.append({"metric":"manhattan","k": k_star.iloc[0]["k"],"log10p":"", "test_error":k_star.iloc[0]["test_error"]})
print(metric_dict)

# Needed for part f
train_k_star = manhattan_error[manhattan_error["train_error"] == manhattan_error["train_error"].min()]
train_dict.append({"metric":"manhattan","k": train_k_star.iloc[0]["k"], "train_error":train_k_star.iloc[0]["train_error"]})
    
```

d. (i) B. with log10(p) âˆˆ {0.1, 0.2, 0.3, . . . , 1}. In this case, use the kâˆ— you found for the Manhattan distance in 1(d)iA. What is the best log10(p)?

```python
p_vals = np.arange(0.1,1.1,0.1)
minkowski_train_error=[]
minkowski_test_error = []

for p in p_vals:
    trainModel = neighbors.KNeighborsClassifier(round(k_star.iloc[0]["k"]), algorithm = "brute", p = 10**p, metric = "minkowski")
    trainModel = trainModel.fit(training_x, training_y)
    trainPreds = trainModel.predict(training_x)
    testPreds = trainModel.predict(test_x)
    minkowski_train_error.append(1 - metrics.classification_report(trainPreds, training_y, output_dict = True).get('accuracy'))
    minkowski_test_error.append(1 - metrics.classification_report(testPreds, test_y, output_dict = True).get('accuracy'))


    
minkowski_error = pd.DataFrame({"p":p_vals, "train_error":minkowski_train_error,"test_error":minkowski_test_error})

p_star = minkowski_error[minkowski_error["test_error"] == minkowski_error["test_error"].min()]
metric_dict.append({"metric":"minkowski","k": k_star.iloc[0]["k"],"log10p":p_star.iloc[0]["p"], "test_error":p_star.iloc[0]["test_error"]})
print(metric_dict)

# Needed for part f
train_k_star = minkowski_error[minkowski_error["train_error"] == minkowski_error["train_error"].min()]
train_dict.append({"metric":"minkowski","train_error":train_k_star.iloc[0]["train_error"]})
```

d. (i) C. which becomes Chebyshev Distance with p â†’ âˆž

```python
k_vals = np.arange(1,197,5)
chebyshev_train_error=[]
chebyshev_test_error = []

for k in k_vals:
    trainModel = neighbors.KNeighborsClassifier(k, metric = "chebyshev")
    trainModel = trainModel.fit(training_x, training_y)
    trainPreds = trainModel.predict(training_x)
    testPreds = trainModel.predict(test_x)
    chebyshev_train_error.append(1 - metrics.classification_report(trainPreds, training_y, output_dict = True).get('accuracy'))
    chebyshev_test_error.append(1 - metrics.classification_report(testPreds, test_y, output_dict = True).get('accuracy'))


    
chebyshev_error = pd.DataFrame({"k":k_vals, "train_error":chebyshev_train_error,"test_error":chebyshev_test_error})

k_star = chebyshev_error[chebyshev_error["test_error"] == chebyshev_error["test_error"].min()]
metric_dict.append({"metric":"chebyshev","k": k_star.iloc[0]["k"],"log10p":"", "test_error":k_star.iloc[0]["test_error"]})
print(metric_dict)

# Needed for part f
train_k_star = chebyshev_error[chebyshev_error["train_error"] == chebyshev_error["train_error"].min()]
train_dict.append({"metric":"chebyshev","k": train_k_star.iloc[0]["k"], "train_error":train_k_star.iloc[0]["train_error"]})
```

d. (ii.) Mahalanobis Distance.

```python
k_vals = np.arange(1,197,5)
mahalanobis_train_error=[]
mahalanobis_test_error = []
training_x = training_set.iloc[:,0:6]

for k in k_vals:
    trainModel = neighbors.KNeighborsClassifier(k, algorithm = "brute", metric = "mahalanobis", metric_params = {'VI': np.linalg.pinv(np.cov(training_x, rowvar=False))})
    trainModel = trainModel.fit(training_x, training_y)
    trainPreds = trainModel.predict(training_x)
    testPreds = trainModel.predict(test_x)
    mahalanobis_train_error.append(1 - metrics.classification_report(trainPreds, training_y, output_dict = True).get('accuracy'))
    mahalanobis_test_error.append(1 - metrics.classification_report(testPreds, test_y, output_dict = True).get('accuracy'))


    
mahalanobis_error = pd.DataFrame({"k":k_vals, "train_error":mahalanobis_train_error,"test_error":mahalanobis_test_error})
k_star = mahalanobis_error[mahalanobis_error["test_error"] == mahalanobis_error["test_error"].min()]
metric_dict.append({"metric":"mahalanobis","k": k_star.iloc[0]["k"],"log10p":"", "test_error":k_star.iloc[0]["test_error"]})
print(metric_dict)

# Needed for part f
train_k_star = mahalanobis_error[mahalanobis_error["train_error"] == mahalanobis_error["train_error"].min()]
train_dict.append({"metric":"mahalanobis","k": train_k_star.iloc[0]["k"], "train_error":train_k_star.iloc[0]["train_error"]})
```

```python
table1 = pd.DataFrame(metric_dict)
table1 = table1.set_index("metric")
display(table1)
```

(e) The majority polling decision can be replaced by weighted decision, in which the
weight of each point in voting is inversely proportional to its distance from the
query/test data point. In this case, closer neighbors of a query point will have
a greater influence than neighbors which are further away. Use weighted voting
with Euclidean, Manhattan, and Chebyshev distances and report the best test
errors when k âˆˆ {1, 6, 11, 16, . . . , 196}.

```python
weighted_dict = []
k_vals = np.arange(1,197,5)

# Euclidian
train_error=[]
test_error = []
for k in k_vals:
    trainModel = neighbors.KNeighborsClassifier(k, weights='distance', metric = "euclidean")
    trainModel = trainModel.fit(training_x, training_y)
    trainPreds = trainModel.predict(training_x)
    testPreds = trainModel.predict(test_x)
    train_error.append(1 - metrics.classification_report(trainPreds, training_y, output_dict = True).get('accuracy'))
    test_error.append(1 - metrics.classification_report(testPreds, test_y, output_dict = True).get('accuracy'))


    
error = pd.DataFrame({"k":k_vals, "train_error":train_error,"test_error":test_error})

k_star = error[error["test_error"] == error["test_error"].min()]
weighted_dict.append({"metric":"euclidian","k": k_star.iloc[0]["k"],"test_error":k_star.iloc[0]["test_error"]})

# Needed for part f
train_k_star = error[error["train_error"] == error["train_error"].min()]
train_dict.append({"metric":"weighted euclidian","k": train_k_star.iloc[0]["k"], "train_error":train_k_star.iloc[0]["train_error"]})


# Manhattan
train_error=[]
test_error = []
for k in k_vals:
    trainModel = neighbors.KNeighborsClassifier(k, weights='distance', metric = "manhattan")
    trainModel = trainModel.fit(training_x, training_y)
    trainPreds = trainModel.predict(training_x)
    testPreds = trainModel.predict(test_x)
    train_error.append(1 - metrics.classification_report(trainPreds, training_y, output_dict = True).get('accuracy'))
    test_error.append(1 - metrics.classification_report(testPreds, test_y, output_dict = True).get('accuracy'))


    
error = pd.DataFrame({"k":k_vals, "train_error":train_error,"test_error":test_error})

k_star = error[error["test_error"] == error["test_error"].min()]
weighted_dict.append({"metric":"manhattan","k": k_star.iloc[0]["k"],"test_error":k_star.iloc[0]["test_error"]})

# Needed for part f
train_k_star = error[error["train_error"] == error["train_error"].min()]
train_dict.append({"metric":"weighted manhattan","k": train_k_star.iloc[0]["k"], "train_error":train_k_star.iloc[0]["train_error"]})

# Chebyshev
train_error=[]
test_error = []
for k in k_vals:
    trainModel = neighbors.KNeighborsClassifier(k, weights='distance', metric = "chebyshev")
    trainModel = trainModel.fit(training_x, training_y)
    trainPreds = trainModel.predict(training_x)
    testPreds = trainModel.predict(test_x)
    train_error.append(1 - metrics.classification_report(trainPreds, training_y, output_dict = True).get('accuracy'))
    test_error.append(1 - metrics.classification_report(testPreds, test_y, output_dict = True).get('accuracy'))


    
error = pd.DataFrame({"k":k_vals, "train_error":train_error,"test_error":test_error})

k_star = error[error["test_error"] == error["test_error"].min()]
weighted_dict.append({"metric":"chebyshev","k": k_star.iloc[0]["k"],"test_error":k_star.iloc[0]["test_error"]})

# Needed for part f
train_k_star = error[error["train_error"] == error["train_error"].min()]
train_dict.append({"metric":"weighted chebyshev","k": train_k_star.iloc[0]["k"], "train_error":train_k_star.iloc[0]["train_error"]})


table2 = pd.DataFrame(weighted_dict)
table2 = table2.set_index("metric")
display(table2)
    
```
