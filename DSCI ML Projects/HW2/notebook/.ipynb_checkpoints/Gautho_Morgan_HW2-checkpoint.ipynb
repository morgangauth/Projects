### 1. (c) For each predictor, fit a simple linear regression model to predict the response. Describe your results. In which of the models is there a statistically significant association between the predictor and the response? Create some plots to back up your assertions. Are there any outliers that you would like to remove from your data for each of these regression tasks?

```python
## Data for 1.(c), response in markdown cell below

y_vals = np.reshape(data["PE"].values, (-1, 1))
fig, axs = plt.subplots(4, 2,  figsize=(10, 15))
fig.suptitle('Linear Regression of Variables vs Energy')
colNames = list(data.columns.values)
coeffs = []
for i in range(4):
    x_vals = np.reshape(data.iloc[:,i].values, (-1, 1))
    x_vals2 = sm.add_constant(x_vals, prepend=False)
    reg = sm.OLS(y_vals, x_vals2).fit()
    coeffs.append(reg.params[0])
    preds = reg.predict(x_vals2)
    cooks_dist = reg.get_influence().cooks_distance 
    outlier_threshold = 4/rows
    
    x_mo = x_vals[cooks_dist[0]<=outlier_threshold]
    x_mo2 = sm.add_constant(x_mo, prepend=False)
    y_mo = y_vals[cooks_dist[0]<=outlier_threshold]
    reg_mo = sm.OLS(y_mo, x_mo2).fit()
    preds_mo = reg_mo.predict(x_mo2)
    
    # Plot linear regression data
    plt.sca(axs[i][0])
    plt.scatter(x_vals, y_vals, color="black", marker="o", s = 0.5)
    plt.plot(x_vals, preds, color="blue", linewidth=3)
    plt.title(colNames[i]+" vs "+ colNames[4])
    plt.xlabel(colNames[i])
    plt.ylabel(colNames[4])
    plt.tight_layout(pad=1.08, h_pad=None, w_pad=None, rect=None)
    
    
    # Plot outlier data
    plt.sca(axs[i][1])
    #outliers
    plt.scatter(x_vals[cooks_dist[0]>outlier_threshold], y_vals[cooks_dist[0]>outlier_threshold], marker="x", color="r", s = 1)
    plt.scatter(x_vals[cooks_dist[0]<=outlier_threshold], y_vals[cooks_dist[0]<=outlier_threshold], color="k", marker="o", s = 0.5)
    plt.plot(x_vals, preds, color="b", linewidth=2)
    plt.plot(x_mo, preds_mo, color="g", linewidth=2)
    plt.title(colNames[i]+" vs "+ colNames[4] + " Outliers")
    plt.legend(['Outliers', 'Non-Outliers', 'Regr w Outliers', 'Regr w/o Outliers'])
    plt.xlabel(colNames[i])
    plt.ylabel(colNames[4])
    
    print("Variable " + colNames[i], ":")
    print(reg.summary(xname = ["const",colNames[i]]))
    print('\n==============================================================================')
    print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
    print('==============================================================================\n')

plt.show()
```

#### 1.(c) Response

Visually, variables 'V' and 'AT' show the strongest correlation between the energy output and the independent variable. However, since all 4 dependent variables have very low P-values (~0.0), they can all be considered statistically significant. 

The plots above show the scatter of the energy output against each independent variable with the linear regression line superimposed on top on the left, and the same plot with outliers highlighted in red on the right. Outliers were identified as any point with a Cook's distance greater than n/4, with n being the number of data points.

### 1. (f) Is there evidence of nonlinear association between any of the predictors and the response? To answer this question, for each predictor X, fit a model of the form Y = Î²0 + Î²1X + Î²2X^2 + Î²3X^3 + e

```python
y_vals = np.reshape(data["PE"].values, (-1, 1))
for i in range(4):
    x_vals = data.iloc[:,i].values
    #raise these to power 1,2,3
    x_vals_2 = np.power(x_vals,2) 
    x_vals_3 = np.power(x_vals,3) 
    x = np.vstack((x_vals,x_vals_2,x_vals_3)).T #alternatively, polynomial features can be used here
    x = sm.add_constant(x, prepend=False)
    summ = sm.OLS(y_vals, x).fit()
    
    print("Variable " + colNames[i], ":")
    print(summ.summary2(xname = [colNames[i], colNames[i]+"^2",colNames[i]+"^3", "const"]))
    
    
    print('\n==============================================================================')
    print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
    print('==============================================================================\n')

```

#### 1.(f) Response
For variable AT, the r squared value increases from 0.899 to 0.912, with a consistently low p-value, indicating a nonlinear association.

For variable V, the r-squared value increased from 0.757 to 0.775, indicating nonlinearity, however some coefficients, such as $V^2$, had a high p-value indicating that this term could be ommitted.

For variable AP, r squared increased from 0.269 to 0.275, with low p-values indicating non-linearity.

For variable RH, r-squared increased from 0.152 to 0.153, with a slightly higher p-value so it does not necessarily indicate non-linearity.

### 1. (g) Is there evidence of association of interactions of predictors with the response? To answer this question, run a full linear regression model with all pairwise interaction terms and state whether any interaction terms are statistically significant.

```python
y_vals = np.reshape(data["PE"].values, (-1, 1))
poly = PolynomialFeatures(interaction_only=True)
x = poly.fit_transform(data.iloc[:,0:4])

x = sm.add_constant(x, prepend=False)
reg = sm.OLS(y_vals, x).fit()
print(reg.summary2(xname=poly.get_feature_names(input_features = colNames)))
```

Train the regression model on a randomly selected 70% subset of the data with all predictors. Test both models on the remaining points and report your train and test MSEs.

### 1. (i) KNN Regression: Perform k-nearest neighbor regression for this dataset using both normalized and raw features. Find the value of k âˆˆ {1, 2, . . . , 100} that gives you the best fit. Plot the train and test errors in terms of 1/k.

```python
# Raw features
k_vals = np.arange(1,101,1)
train_error = []
test_error = []
for k in k_vals:
    neighReg = neighbors.KNeighborsRegressor(n_neighbors=k)
    neighReg = neighReg.fit(x_train, y_train)
    trainPreds = neighReg.predict(x_train)
    testPreds = neighReg.predict(x_test)
    train_error.append(metrics.mean_squared_error(y_train, trainPreds))
    test_error.append(metrics.mean_squared_error(y_test, testPreds))

error = pd.DataFrame({"k":k_vals, "train_error":train_error,"test_error":test_error})
k_star = error[error["test_error"] == error["test_error"].min()]
print("The value of k which minimizes the test error is %d with a test error rate of %.2f" % (k_star.iloc[0]["k"],k_star.iloc[0]["test_error"] ), "%.")
    

# Plot
plt.figure(figsize=(5,3)) 
plt.plot(1/error["k"], error["train_error"], label = "Training Error")
plt.plot(1/error["k"], error["test_error"], label = "Test Error")
plt.xlabel("1/k value")
plt.ylabel("Error Rate")
plt.title("Test vs. Training Error Rate for Raw Data")
plt.legend()
plt.show()

all_mse_train.append(error["train_error"].min())
all_mse_test.append(error["test_error"].min())
all_models.append("raw_knn")
all_adj_r.append("")
```

```python
# Normalized features
k_vals = np.arange(1,101,1)
train_error = []
test_error = []
scaler = preprocessing.StandardScaler().fit(x_train)
x_train_normalized = scaler.transform(x_train)
x_test_normalized = scaler.transform(x_test) 

for k in k_vals:
    neighReg = neighbors.KNeighborsRegressor(n_neighbors=k)
    neighReg = neighReg.fit(x_train_normalized, y_train)
    trainPreds = neighReg.predict(x_train_normalized)
    testPreds = neighReg.predict(x_test_normalized)
    train_error.append(metrics.mean_squared_error(y_train, trainPreds))
    test_error.append(metrics.mean_squared_error(y_test, testPreds))

error = pd.DataFrame({"k":k_vals, "train_error":train_error,"test_error":test_error})
k_star = error[error["test_error"] == error["test_error"].min()]
print("The value of k which minimizes the test error is %d with a test error rate of %.2f" % (k_star.iloc[0]["k"],k_star.iloc[0]["test_error"] ), "%.")
    

# Plot
plt.figure(figsize=(5,3)) 
plt.plot(1/error["k"], error["train_error"], label = "Training Error")
plt.plot(1/error["k"], error["test_error"], label = "Test Error")
plt.xlabel("1/k value")
plt.ylabel("Error Rate")
plt.title("Test vs. Training Error Rate for Normalized Data")
plt.legend()
plt.show()

all_mse_train.append(error["train_error"].min())
all_mse_test.append(error["test_error"].min())
all_models.append("normalized_knn")
all_adj_r.append("")
```

### 1. (j) Compare the results of KNN Regression with the linear regression model that has the smallest test error and provide your analysis.

```python
results = pd.DataFrame({"mse_test":all_mse_test,
       "mse_train":all_mse_train,
       "model":all_models,
       "adj_r_squ":all_adj_r})
results = results.set_index('model')
display(results)
```

#### 1. (j) Response
As can be seen in the above grid, the smallest test error was found for Normalized KNN regression. Of the linear regression models, the one with the lowest error was the model with the interactions and quadratic features included, which is contrary to the expectation that the model excluding the insignificant features would have the lowest errors. However, the interactions excluding the insignificant features _does_ have the highest adjusted R-squared value, which is as expected since it should be a closer fit.

1. For each of parts (a) through (d), indicate whether we would generally
expect the performance of a flexible statistical learning method to be
better or worse than an inflexible method. Justify your answer.<br>
(a) The sample size n is extremely large, and the number of predictors p is small.<br>
(b) The number of predictors p is extremely large, and the number
of observations n is small.<br>
(c) The relationship between the predictors and response is highly
non-linear.<br>
(d) The variance of the error terms, i.e. Ïƒ2 = Var(Ïµ), is extremely
high.<br>

**(a)** For a large number of samples with few features, an flexible method would perform better because the large number of samples would need a flexible model to fit the data. <br>
**(b)** For many different predictors with a small dataset, an inflexible model would perform better. Too much flexibility would likely lead to overfitting. <br>
**(c)** For a non-linear relationship between predictors and response, a flexible model would perform better as it will have more degrees of freedom to fit a non-linear shape.<br>
**(d)** With very high variance of the error terms, an inflexible model would work better as it would likely overfit the error terms and the high variance.

## 3. ISLR: 2.4.7

7. The table below provides a training data set containing six observations, three predictors, and one qualitative response variable. <br>
Obs. X1 X2 X3 Y<br>
1 0 3 0 Red<br>
2 2 0 0 Red<br>
3 0 1 3 Red<br>
4 0 1 2 Green<br>
5 âˆ’1 0 1 Green<br>
6 1 1 1 Red<br>
Suppose we wish to use this data set to make a prediction for Y when
X1 = X2 = X3 = 0 using K-nearest neighbors.<br>
(a) Compute the Euclidean distance between each observation and
the test point, X1 = X2 = X3 = 0.<br>
(b) What is our prediction with K = 1? Why?<br>
(c) What is our prediction with K = 3? Why?<br>
(d) If the Bayes decision boundary in this problem is highly nonlinear, then would we expect the best value for K to be large or
small? Why?<br>

**(a)** The euclidian distance can be found using Pythagorean's theorem:

$\sqrt{(x1 - x1')^2 + (x2 - x2')^2 + (x3 - x3')^2}$

Obs 1: $\sqrt{(0 - 0)^2 + (3 - 0)^2 + (0 - 0)^2} = 3$ <br>
Obs 2: $\sqrt{(2 - 0)^2 + (0 - 0)^2 + (0 - 0)^2} = 2$ <br>
Obs 3: $\sqrt{(0 - 0)^2 + (1 - 0)^2 + (3 - 0)^2} = \sqrt{10}$ <br>
Obs 4: $\sqrt{(0 - 0)^2 + (1 - 0)^2 + (2 - 0)^2} = \sqrt{5}$ <br>
Obs 5: $\sqrt{(-1 - 0)^2 + (0 - 0)^2 + (1 - 0)^2} = \sqrt{2}$ <br>
Obs 6: $\sqrt{(1 - 0)^2 + (1 - 0)^2 + (1 - 0)^2} = \sqrt{3}$ <br>

**(b)** For k=1, the classifier will only consider the 1-closest neighbor, In this case, Observation 5 has the nearest distance and so the prediction will be "Green". <br>

**(c)** With k=3, the classifier will consider the 3 nearest neighbors. In this case, the nearest neighbors are Observation 5, Observation 6, and Observation 2. The observed classifications are Green, Red, and Red, respectively. The probability that the test point is Red is 2/3, so the prediction will be "Red". <br>

**(d)** With a highly nonlinear boundary, a smaller k value would be better than a large one, since a smaller k value allows for a more flexible/non-linear boundary. A larger k would consider more nearby data points and the boundary would approach a more linear shape.
