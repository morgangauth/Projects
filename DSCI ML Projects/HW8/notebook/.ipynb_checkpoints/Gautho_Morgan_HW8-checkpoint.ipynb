##  1. Supervised, Semi-Supervised, and Unsupervised Learning

### (a) Download the Breast Cancer Wisconsin (Diagnostic) Data Set from: https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29. Download the data in https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data, which has IDs, classes (Benign=B, Malignant=M), and 30 attributes. This data has two output classes.

```python
colNames = ["ID", "Diagnosis",  "meanradius", "meantexture", "meanperimeter", "meanarea" , 
            "meansmoothness", "meancompactness", "meanconcavity", "meanconcavepoints" ,"meansymmetry", "meanfractaldimension",
           "seradius", "setexture", "seperimeter", "searea" , 
            "sesmoothness", "secompactness", "seconcavity", "seconcavepoints" ,"sesymmetry", "sefractaldimension",
            "worstradius", "worsttexture", "worstperimeter", "worstarea" , 
            "worstsmoothness", "worstcompactness", "worstconcavity", "worstconcavepoints" ,"worstsymmetry", "worstfractaldimension"]
allData = pd.read_csv("../data/wdbc.data", names = colNames)
allData['class'] = np.where(allData['Diagnosis']== 'M', 1, 0)
allData.head()
```

### (b) Monte-Carlo Simulation: Repeat the following procedures for supervised, unsupervised, and semi-supervised learning M = 30 times, and use randomly selected train and test data (make sure you use 20% of both the positve and negative classes as the test set). Then compare the average scores (accuracy, precision, recall, F1-score, and AUC) that you obtain from each algorithm.


```python
x_data = allData.iloc[:,2:-1]
y_data = allData.iloc[:,-1]

m_data = allData[allData.Diagnosis == "M"]
b_data = allData[allData.Diagnosis == "B"]
m_train, m_test = train_test_split(m_data, test_size = 0.2)
b_train, b_test = train_test_split(b_data, test_size = 0.2)
trainData = pd.concat([m_train, b_train])
testData = pd.concat([m_test, b_test])

x_train = trainData.iloc[:,2:-1]
x_test = testData.iloc[:,2:-1]
y_train = trainData.iloc[:,-1]
y_test = testData.iloc[:,-1]

m_vals = np.arange(1,31)
```

#### i. Supervised Learning: Train an L1-penalized SVM to classify the data. Use 5 fold cross validation to choose the penalty parameter. Use normalized data. Report the average accuracy, precision, recall, F1-score, and AUC, for both training and test sets over your M runs. Plot the ROC and report the confusion matrix for training and testing in one of the runs.

```python
training_results1 = []
test_results1 = []
for m in m_vals:
    X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, stratify=y_data)
    min_max_scaler = preprocessing.MinMaxScaler()
    X_train = min_max_scaler.fit_transform(X_train)
    X_test = min_max_scaler.fit_transform(X_test)
    c_vals = np.logspace(-5, 8, 10)
    params = {'C': c_vals}
    supervised_model = LinearSVC(penalty= 'l1', dual=False, max_iter = 5000)
    gridCV = GridSearchCV(supervised_model, param_grid=params, cv=5)
    gridCV.fit(X_train, y_train)
    
    # training results
    y_train_pred = gridCV.predict(X_train)
    y_train_scores = gridCV.decision_function(X_train)
    
    train_accuracy = accuracy_score(y_train, y_train_pred)
    train_precision = precision_score(y_train, y_train_pred)
    train_recall = recall_score(y_train, y_train_pred)
    train_f1 = f1_score(y_train, y_train_pred)
    train_auc = roc_auc_score(y_train,y_train_pred)
    
    training_results1.append({"accuracy":train_accuracy, "precision":train_precision, "recall":train_recall,
                         "f1score":train_f1, "auc":train_auc})
    
    # test results
    y_test_pred = gridCV.predict(X_test)
    y_test_scores = gridCV.decision_function(X_test)
    
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_precision = precision_score(y_test, y_test_pred)
    test_recall = recall_score(y_test, y_test_pred)
    test_f1 = f1_score(y_test, y_test_pred)
    test_auc = roc_auc_score(y_test,y_test_pred)
    
    test_results1.append({"accuracy":test_accuracy, "precision":test_precision, "recall":test_recall,
                         "f1score":test_f1, "auc":test_auc})
    
    #print results
    print("Iteration ", m,": ")
    print("Best C = ", gridCV.best_params_)
    print("Training results: ", training_results1[m-1])
    print("Test results: ", test_results1[m-1])
    
    
    #ROC and Confusion Matrix:
    if m == 1:
        #Train
        mat = confusion_matrix(y_train,y_train_pred)
        tn, fp, fn, tp = mat.ravel()
        true_positive_rate = tp / (tp + fn)
        true_negative_rate = tn / (tn + fp)
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f_score = 2* ((precision * recall) / (precision + recall))

        sns.set(font_scale=1.4)
        sns.heatmap(mat,square= True, annot=True, cbar= False, fmt='g')
        plt.xlabel("Predicted B     Predicted M")
        plt.ylabel("Actual M     Actual B")
        plt.title("Training Confusion Matrix")
        print(mat)
        plt.show()

        fpr, tpr, thresholds = roc_curve(y_train, y_train_scores)
        roc_auc = auc(fpr,tpr)
        plt.plot(fpr, tpr, color='darkorange', label='AUC = %0.2f' % roc_auc)
        plt.plot([0, 1], [0, 1], color='black',  linestyle=':')
        plt.xlim([-0.1, 1.1])
        plt.ylim([-0.1, 1.1])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Training ROC')
        plt.show()
        
        #Test
        mat = confusion_matrix(y_test,y_test_pred)
        tn, fp, fn, tp = mat.ravel()
        true_positive_rate = tp / (tp + fn)
        true_negative_rate = tn / (tn + fp)
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f_score = 2* ((precision * recall) / (precision + recall))

        sns.set(font_scale=1.4)
        sns.heatmap(mat,square= True, annot=True, cbar= False, fmt='g')
        plt.xlabel("Predicted B     Predicted M")
        plt.ylabel("Actual M     Actual B")
        plt.title("Test Confusion Matrix")
        print(mat)
        plt.show()

        fpr, tpr, thresholds = roc_curve(y_test, y_test_scores)
        roc_auc = auc(fpr,tpr)
        plt.plot(fpr, tpr, color='darkorange', label='AUC = %0.2f' % roc_auc)
        plt.plot([0, 1], [0, 1], color='black',  linestyle=':')
        plt.xlim([-0.1, 1.1])
        plt.ylim([-0.1, 1.1])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Test ROC')
        plt.show()
        
        
```

### iii. Unsupervised Learning: Run k-means algorithm on the whole training set. Ignore the labels of the data, and assume k = 2.

#### A. Run the k-means algorithm multiple times. Make sure that you initialize the algoritm randomly. How do you make sure that the algorithm was not trapped in a local minimum?
#### B. Compute the centers of the two clusters and find the closest 30 datapoints to each center. Read the true labels of those 30 data points andtake a majority poll within them. The majority poll becomes the labelpredicted by k-means for the members of each cluster. Then compare thelabels provided by k-means with the true labels of the training data andreport the average accuracy, precision, recall, F1-score, and AUC over Mruns, and ROC and the confusion matrix for one of the runs.
#### C. Classify test data based on their proximity to the centers of the clusters. Report the average accuracy, precision, recall, F1-score, and AUC over M runs, and ROC and the confusion matrix for one of the runs for the test data.2


```python
training_results3 = []
test_results3 = []
for m in m_vals: 
    #split train/test
    X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, stratify=y_data)
    min_max_scaler = preprocessing.MinMaxScaler()
    X_train = min_max_scaler.fit_transform(X_train)
    X_test = min_max_scaler.fit_transform(X_test)
    kmeans = KMeans(n_clusters=2,init='random')
    kmeans.fit(X_train)
    labels = kmeans.labels_
    
    center1 = kmeans.cluster_centers_[0]
    center2 = kmeans.cluster_centers_[1]

    X_train = pd.DataFrame(X_train)    
    X_train["distanceC1"] = X_train.apply(lambda row: np.linalg.norm(center1 - row), axis = 1)
    X_train["distanceC2"] = X_train.apply(lambda row: np.linalg.norm(center2 - row.iloc[0:-1]), axis = 1)
    
    trainData = X_train
    trainData["diagnostic"] = y_train
    #trainData = trainData.fillna(0)
    
    #find 30 closest
    c1 = trainData.sort_values(by='distanceC1').iloc[:30]['diagnostic'].value_counts().index[0]
    c2 = trainData.sort_values(by='distanceC2').iloc[:30]['diagnostic'].value_counts().index[0]
    
    
    #majority poll 
    trainData["labels"] = labels
    trainData["predLabel"] = trainData.apply(lambda x: c1 if x['labels'] == 0 else c2, axis=1)
    
    # training results
    y_train = trainData["diagnostic"]
    y_train_pred = trainData["predLabel"]
    
    y_train = y_train.fillna(1)
    
    train_accuracy = accuracy_score(y_train, y_train_pred)
    train_precision = precision_score(y_train, y_train_pred)
    train_recall = recall_score(y_train, y_train_pred)
    train_f1 = f1_score(y_train, y_train_pred)
    train_auc = roc_auc_score(y_train,y_train_pred)
    
    training_results3.append({"accuracy":train_accuracy, "precision":train_precision, "recall":train_recall,
                         "f1score":train_f1, "auc":train_auc})
    
    # test results
    
    X_test = pd.DataFrame(X_test)    
    X_test["distanceC1"] = X_test.apply(lambda row: np.linalg.norm(center1 - row), axis = 1)
    X_test["distanceC2"] = X_test.apply(lambda row: np.linalg.norm(center2 - row.iloc[0:-1]), axis = 1)
    testData = X_test
    testData["diagnostic"] = y_test
    testData["predLabel"] = testData.apply(lambda x: c1 if x['distanceC1'] < x['distanceC2'] else c2, axis=1)
    
    testData = testData.fillna(0)
    
    # test results
    y_test = testData["diagnostic"]
    y_test_pred = testData["predLabel"]
    
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_precision = precision_score(y_test, y_test_pred)
    test_recall = recall_score(y_test, y_test_pred)
    test_f1 = f1_score(y_test, y_test_pred)
    test_auc = roc_auc_score(y_test,y_test_pred)
    
    test_results3.append({"accuracy":test_accuracy, "precision":test_precision, "recall":test_recall,
                         "f1score":test_f1, "auc":test_auc})
    
    #print results
    print("Iteration ", m,": ")
    print("Training results: ", training_results3[m-1])
    print("Test results: ", test_results3[m-1])
    
    
    #ROC and Confusion Matrix:
    if m == 1:
        #Train
        mat = confusion_matrix(y_train,y_train_pred)
        tn, fp, fn, tp = mat.ravel()
        true_positive_rate = tp / (tp + fn)
        true_negative_rate = tn / (tn + fp)
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f_score = 2* ((precision * recall) / (precision + recall))

        sns.set(font_scale=1.4)
        sns.heatmap(mat,square= True, annot=True, cbar= False, fmt='g')
        plt.xlabel("Predicted B     Predicted M")
        plt.ylabel("Actual M     Actual B")
        plt.title("Training Confusion Matrix")
        print(mat)
        plt.show()

        fpr, tpr, thresholds = roc_curve(y_train, y_train_pred)
        roc_auc = auc(fpr,tpr)
        plt.plot(fpr, tpr, color='darkorange', label='AUC = %0.2f' % roc_auc)
        plt.plot([0, 1], [0, 1], color='black',  linestyle=':')
        plt.xlim([-0.1, 1.1])
        plt.ylim([-0.1, 1.1])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Training ROC')
        plt.show()
        
        #Test
        mat = confusion_matrix(y_test,y_test_pred)
        tn, fp, fn, tp = mat.ravel()
        true_positive_rate = tp / (tp + fn)
        true_negative_rate = tn / (tn + fp)
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f_score = 2* ((precision * recall) / (precision + recall))

        sns.set(font_scale=1.4)
        sns.heatmap(mat,square= True, annot=True, cbar= False, fmt='g')
        plt.xlabel("Predicted B     Predicted M")
        plt.ylabel("Actual M     Actual B")
        plt.title("Test Confusion Matrix")
        print(mat)
        plt.show()

        fpr, tpr, thresholds = roc_curve(y_test, y_test_pred)
        roc_auc = auc(fpr,tpr)
        plt.plot(fpr, tpr, color='darkorange', label='AUC = %0.2f' % roc_auc)
        plt.plot([0, 1], [0, 1], color='black',  linestyle=':')
        plt.xlim([-0.1, 1.1])
        plt.ylim([-0.1, 1.1])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Test ROC')
        plt.show()
        
        
```

Use RBF kernel with gamma=1 or find a gamma for which the two clutsres have the same balance as the one in original data set (if the positive class has p and the negative class has n samples, the two clusters must have p and n members). Do not label data based on their proximity to cluster center, because spectral clustering may give you non-convex clusters . Instead, use fit âˆ’ predict method.



```python
training_results4 = []
test_results4 = []
for m in m_vals: 
    #split train/test
    X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, stratify=y_data)
    min_max_scaler = preprocessing.MinMaxScaler()
    X_train = min_max_scaler.fit_transform(X_train)
    X_test = min_max_scaler.fit_transform(X_test)
    
    spectral = SpectralClustering(n_clusters=2, gamma=1.0, affinity='rbf')
    labels = spectral.fit_predict(X_train)
    
    X_train = pd.DataFrame(X_train)
    X_train["labels"] = labels
    trainData = X_train
    trainData["class"] = y_train.values
    
    ## Label
    for i in range (0,2):
        major = trainData[trainData.labels == i]['class'].value_counts().index[0]
        trainData.loc[trainData.labels == i, 'pred'] = major
    trainData['pred'] = trainData['pred'].astype(int)

    # training results
    
    y_train = trainData["class"]
    y_train_pred = trainData["pred"]
    
    
    train_accuracy = accuracy_score(y_train, y_train_pred)
    train_precision = precision_score(y_train, y_train_pred)
    train_recall = recall_score(y_train, y_train_pred)
    train_f1 = f1_score(y_train, y_train_pred)
    train_auc = roc_auc_score(y_train,y_train_pred)
    
    training_results4.append({"accuracy":train_accuracy, "precision":train_precision, "recall":train_recall,
                         "f1score":train_f1, "auc":train_auc})
    
    # test results
    
    X_test = pd.DataFrame(X_test)   
    y_test = pd.DataFrame(y_test)
    labels = spectral.fit_predict(X_test)
    X_test["labels"] = labels
    testData = X_test
    
    
    testData["class"] = y_test.values
    
    
    # test results
    for i in range (0,2):
        major = testData[testData.labels == i]['class'].value_counts().index[0]
        testData.loc[testData.labels == i, 'pred'] = major
    testData['pred'] = testData['pred'].astype(int)

    # training results
    y_test = testData["class"]
    y_test_pred = testData["pred"]
    
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_precision = precision_score(y_test, y_test_pred)
    test_recall = recall_score(y_test, y_test_pred)
    test_f1 = f1_score(y_test, y_test_pred)
    test_auc = roc_auc_score(y_test,y_test_pred)
    
    test_results4.append({"accuracy":test_accuracy, "precision":test_precision, "recall":test_recall,
                         "f1score":test_f1, "auc":test_auc})
    
    #print results
    print("Iteration ", m,": ")
    print("Training results: ", training_results4[m-1])
    print("Test results: ", test_results4[m-1])
    
    
    #ROC and Confusion Matrix:
    if m == 1:
        #Train
        mat = confusion_matrix(y_train,y_train_pred)
        tn, fp, fn, tp = mat.ravel()
        true_positive_rate = tp / (tp + fn)
        true_negative_rate = tn / (tn + fp)
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f_score = 2* ((precision * recall) / (precision + recall))

        sns.set(font_scale=1.4)
        sns.heatmap(mat,square= True, annot=True, cbar= False, fmt='g')
        plt.xlabel("Predicted B     Predicted M")
        plt.ylabel("Actual M     Actual B")
        plt.title("Training Confusion Matrix")
        print(mat)
        plt.show()

        fpr, tpr, thresholds = roc_curve(y_train, y_train_pred)
        roc_auc = auc(fpr,tpr)
        plt.plot(fpr, tpr, color='darkorange', label='AUC = %0.2f' % roc_auc)
        plt.plot([0, 1], [0, 1], color='black',  linestyle=':')
        plt.xlim([-0.1, 1.1])
        plt.ylim([-0.1, 1.1])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Training ROC')
        plt.show()
        
        #Test
        mat = confusion_matrix(y_test,y_test_pred)
        tn, fp, fn, tp = mat.ravel()
        true_positive_rate = tp / (tp + fn)
        true_negative_rate = tn / (tn + fp)
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f_score = 2* ((precision * recall) / (precision + recall))

        sns.set(font_scale=1.4)
        sns.heatmap(mat,square= True, annot=True, cbar= False, fmt='g')
        plt.xlabel("Predicted B     Predicted M")
        plt.ylabel("Actual M     Actual B")
        plt.title("Test Confusion Matrix")
        print(mat)
        plt.show()

        fpr, tpr, thresholds = roc_curve(y_test, y_test_pred)
        roc_auc = auc(fpr,tpr)
        plt.plot(fpr, tpr, color='darkorange', label='AUC = %0.2f' % roc_auc)
        plt.plot([0, 1], [0, 1], color='black',  linestyle=':')
        plt.xlim([-0.1, 1.1])
        plt.ylim([-0.1, 1.1])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Test ROC')
        plt.show()
```

## 2. Active Learning Using Support Vector Machines

### (a) Download the banknote authentication Data Set from: https://archive.ics.uci.edu/ml/datasets/banknote+authentication. Choose 472 data points randomly as the test set, and the remaining 900 points as the training set. This is a binary classification problem.


```python
colNames = ["variance","skewness","curtosis","entropy","class"]
bankData = pd.read_csv("../data/data_banknote_authentication.txt", names = colNames)
bankData.head()

trainData, testData = train_test_split(bankData, test_size = 472)
m_vals = np.arange(1,51)
x_test = testData.iloc[:,:-1]
y_test = testData.iloc[:,-1]
```

### (b) Repeat each of the following two procedures 50 times. You will have 50 errors for 90 SVMs per each procedure

#### i. Train a SVM with a pool of 10 randomly selected data points from the trainingset using linear kernel and L1 penalty. Select the penalty parameter using5-fold cross validation.4 Repeat this process by adding 10 other randomly selected data points to the pool, until you use all the 900 points. Do NOTreplace the samples back into the training set at each step. Calculate thetest error for each SVM. You will have 90 SVMs that were trained using 10,20, 30, ... , 900 data points and their 90 test errors. You have implemented passive learning.


```python
errors = pd.DataFrame(columns=list(range(10, 910, 10)))

for m in m_vals:
    selected = pd.DataFrame()
    trainingSet = trainData
    numPoints = 0
    iterErrors = []
    
    while numPoints < 901:
        if numPoints == 890:
            selected = pd.concat([selected, trainingSet], axis = 0)
            numPoints += 10
        else :
            remaining, addedVals = train_test_split(trainingSet, test_size = 10)
            trainingSet = remaining
            selected = pd.concat([selected, addedVals], axis = 0)
            numPoints += 10
        x_train = selected.iloc[:,:-1]
        y_train = selected.iloc[:,-1].astype("int")
        
        params = {'C':np.logspace(-5, 8, 10)}
        model = LinearSVC(penalty='l1', dual = False)
        gridCV = GridSearchCV(model, param_grid=params, cv=5, scoring = 'accuracy')  
        gridCV.fit(x_train, y_train)
        iterErrors.append(1 - gridCV.score(x_test,y_test))

    errors = errors.append(pd.Series(iterErrors), ignore_index=True)


    
    print("Iteration ",m,": ")
    print(iterErrors,"\n")
    
        
```

#### ii. Train a SVM with a pool of 10 randomly selected data points from the training set5 using linear kernel and L1 penalty. Select the parameters of the SVM with 5-fold cross validation. Choose the 10 closest data points in the training set to the hyperplane of the SVM6 and add them to the pool. Do not replace the samples back into the training set. Train a new SVM using the pool. Repeat this process until all training data is used. You will have 90 SVMs that were trained using 10, 20, 30,..., 900 data points and their 90 test errors. You have implemented active learning.

```python
errorsActive = pd.DataFrame(columns=list(range(10, 910, 10)))

for m in m_vals:
    selected = pd.DataFrame()
    trainingSet = trainData
    numPoints = 0
    iterErrors = []
    
    remaining, addedVals = train_test_split(trainingSet, test_size = 10)
    trainingSet = remaining
    selected = pd.concat([selected, addedVals], axis = 0)
    numPoints += 10
    x_train = selected.iloc[:,:-1]
    y_train = selected.iloc[:,-1].astype("int")
        
    params = {'C':np.logspace(-5, 8, 10)}
    model = LinearSVC(penalty='l1', dual = False)
    gridCV = GridSearchCV(model, param_grid=params, cv=5, scoring = 'accuracy')  
    gridCV.fit(x_train, y_train)
    iterErrors.append(1 - gridCV.score(x_test,y_test))

    for i in range(0, 89):
        trainingSet["Distance"] = np.abs(gridCV.decision_function(trainingSet.iloc[:,:4]))
        trainingSet = trainingSet.sort_values(by=["Distance"])
        
        addedVals = trainingSet.iloc[0:10,0:5]
        selected = pd.concat([selected, addedVals], axis = 0)
        trainingSet = trainingSet.iloc[10:,0:5]
        x_train = selected.iloc[:,:-1]
        y_train = selected.iloc[:,-1].astype("int")
        
        #retrain
        gridCV.fit(x_train, y_train)
        iterErrors.append(1 - gridCV.score(x_test,y_test))
    
    
    
    
    errorsActive = errorsActive.append(pd.Series(iterErrors), ignore_index=True)


    
    print("Iteration ",m,": ")
    print(iterErrors,"\n")
    
        
```

### (c) Average the 50 test errors for each of the incrementally trained 90 SVMs in 2(b)i and 2(b)ii. By doing so, you are performing a Monte Carlo simulation. Plot average test error versus number of training instances for both active and passive learners on the same figure and report your conclusions. Here, you are actually obtaining a learning curve by Monte-Carlo simulation

```python
#Averages
print("Passive Learning:")
passive = errors.describe().loc[['mean']]
display(passive)
print("Active Learning:")
active = errorsActive.describe().loc[['mean']]
display(active)
```
